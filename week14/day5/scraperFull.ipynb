```python
#scraper
from bs4 import BeautifulSoup
import requests
import json
import time

start=time.time()
number_of_pages=10
category='environnement'    
articles=[]

#Loop to iterate over each request to the website for each page number
for i in range(1,number_of_pages+1): 
    url=f'https://lexpress.mu/societe/{category}?page={i}'

    my_page=requests.get(url)
    page=BeautifulSoup(my_page.content,'html.parser')
    section=page.findAll(class_='flex flex-col gap-4')[1]
    selected_articles=section.find_all('a')
    for article in selected_articles:
        try:
            title=article.find(class_='text-red-600').text
            description=article.find(class_='mb-2 text-xl font-bold font-serif').text
            img="https://lexpress.mu"+article.find('img')['src']
            link="https://lexpress.mu"+article['href']

            timeDate=article.find(class_='mt-4 text-gray-500 text-sm flex flex-row gap-2 items-center')
            date=timeDate.findAll('span')[1].text
            time=timeDate.findAll('span')[3].text

            new_article={
                    'title':title,
                    'link':link,
                    'img':img,
                    'description': description,
                    'date': date,
                    'time': time
            }
            articles.append(new_article)
        except:
            pass

end=time.time()
duration=end-start
print(f"Time taken to process:{duration}")
```

```python
#export articles to a padas dataframe
import pandas as pd
import numpy as np

df=pd.DataFrame(articles)
df
```

```python
#export to csv or excel
df.to_csv('lexpress.csv',index=False, encoding='utf-8')
```
