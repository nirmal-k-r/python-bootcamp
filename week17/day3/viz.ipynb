```python
df = pd.DataFrame({
    'name':['john','mary','peter','jeff','bill','lisa','jose'],
    'age':[23,78,22,19,45,33,20],
    'gender':['M','F','M','M','M','F','M'],
    'state':['california','dc','california','dc','california','texas','texas'],
    'num_children':[2,0,0,3,2,1,4],
    'num_pets':[5,1,0,5,2,2,3]
})

df.head()
```

```python
df.plot(kind='scatter',x='num_children',y='num_pets',color='red')
```

```python
plt.scatter(df['age'],df['num_children'],color='blue')
plt.title('Age vs Number of Children')
plt.xlabel('Age')
plt.ylabel('Number of Children')
```

```python
df
```

```python
x=['Female','Male']
y=df.groupby('gender')['num_children'].sum()

print(y)
fig,ax=plt.subplots()
ax.barh(x,y)
ax.set_ylabel('Num of children')
ax.set_xlabel('Gender')


plt.title('Male vs female num of children')
plt.show()
```

```python
fig,ax=plt.subplots()
fig.set_size_inches(10,5) # Set the size of the plot
ax.pie(y,labels=x)

ax.set_title('Percentage of children by gender')
plt.show()

fig.savefig('plot.png')
```

```python
x = np.linspace(0,100,11)
y=(2*x+5)/5

fig,ax=plt.subplots()
fig.set_size_inches(7,5)
ax.plot(x,y,color='red',marker='o',linestyle='dashed',linewidth=2,markersize=5)
plt.show()
```

```python
x=['Female','Male']
y=df.groupby('gender')['num_children'].sum()

print(y)
fig,ax=plt.subplots(1,2) # Create a 1 row, 2 column plot
fig.set_size_inches(10,5) # Set the size of the plot
ax[0].bar(x,y)
ax[0].set_ylabel('Num of children')
ax[0].set_xlabel('Gender')

ax[1].pie(y,labels=x)
ax[1].legend(loc='lower right',shadow=True)


plt.title('Male vs female num of children')
plt.show()
```

```python
#annoate plot
x = np.linspace(0,100,11)
y=(2*x+5)/5


fig,ax=plt.subplots()
fig.set_size_inches(7,5)
ax.set_title('Annotated Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.plot(x,y)


# Annotate the plot at pt 40,17
ax.annotate('perfect point',(40,17),xytext=(45,13),arrowprops=dict(facecolor='black',shrink=0.01))

plt.show()
```

```python
x = np.linspace(0,100,11)
y=(2*x+5)/5

fig,ax=plt.subplots()
ax.errorbar(x,y,xerr=10)
plt.show()
```

```python
#annoate plot
x = np.linspace(0,100,11)
y=(2*x+5)/5


fig,ax=plt.subplots()
fig.set_size_inches(7,5)
ax.set_title('Annotated Scatter Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.plot(x,y,color='blue')

ax2=ax.twinx() # Create a second y-axis

x=x*2
y=2*x+50
ax2.plot(x,y,color='red')

labels=['Line 2','Line1']
plt.legend(labels,loc='upper left')

plt.show()
```
