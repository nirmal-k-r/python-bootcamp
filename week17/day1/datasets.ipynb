```python
#convert types
df[['Engine']]=df[['Engine']].astype('float')
df.info()
```

```python
#percentage of manual cars above 2000 cc -> bar chart
manual_cars=df.query('Transmission=="Manual" and Engine>2000')

num_manual_cars_above_2000cc=manual_cars.shape[0]
total_cars=df.shape[0]
percentage=(num_manual_cars_above_2000cc/total_cars)*100
print(percentage)

results=pd.Series([percentage,100-percentage],index=['Manual','Automatic'])
results.plot(title='Charts',kind='pie')
```
